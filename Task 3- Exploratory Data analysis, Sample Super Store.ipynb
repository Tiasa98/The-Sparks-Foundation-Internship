```python
import numpy  as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data=pd.read_csv('C:\\Users\\LENOVO\\Downloads\\SampleSuperstore.csv')
```

```python
data.head(10)
```

# Data Preparation

```python
data.describe()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
#Check for dublicate values
data.duplicated().sum()
```

```python
#Remove those dublicate values
data.drop_duplicates(inplace=True)
```

# State wise sales distribution

```python
fig,ax=plt.subplots(figsize=(30,12))
x=data['State']
y=data['Sales']
ax.set_xlabel('State')
ax.set_ylabel('Sales')
ax.set_title('State wise Sales distribution')

fig.autofmt_xdate()
ax.bar(x,y)
plt.show()
```

We can see Florida has the highest sales and Kansas has lowest sales.

# State wise profit distribution

```python
fig,ax=plt.subplots(figsize=(30,12))
x=data['State']
y=data['Profit']
ax.set_xlabel('State')
ax.set_ylabel('Profit')
ax.set_title('State wise Profit distribution')

fig.autofmt_xdate()
ax.bar(x,y)
plt.show()
```

We can see though every state has some amount of sales but some states has to face negetive profit. We have to take care of this.

# Region wise sales distribution

```python
fig,ax=plt.subplots(figsize=(30,12))
x=data['Region']
y=data['Sales']
ax.set_xlabel('Region')
ax.set_ylabel('Sales')
ax.set_title('Region wise Sales distribution')

fig.autofmt_xdate()
ax.bar(x,y)
plt.show()
```

We can see South region has most amount of sales and East has lowest.

# Region wise profit distribution

```python
fig,ax=plt.subplots(figsize=(30,12))
x=data['Region']
y=data['Profit']
ax.set_xlabel('Region')
ax.set_ylabel('Profit')
ax.set_title('Region wise Profit distribution')

fig.autofmt_xdate()
ax.bar(x,y)
plt.show()
```

Surprising, though South region has most of sales but Central has most of the profits. For East region, it has lowest sales, lowest profit.

# Region wise category classification

```python
plt.figure(figsize = (30, 12))
sns.countplot(x = 'Category', hue = 'Region', data = data, palette = 'rocket')
plt.xticks(rotation = 45, fontsize = 20)
plt.yticks(fontsize = 20)
plt.xlabel('Categories', fontsize = 25)
plt.ylabel('', fontsize = 20)
plt.legend(loc = 5, fontsize = 20)
```

# Category wise sales distribution

```python
fig,ax=plt.subplots(figsize=(30,12))
x=data['Category']
y=data['Sales']
ax.set_xlabel('Category')
ax.set_ylabel('Sales')
ax.set_title('Category wise Sales distribution')

fig.autofmt_xdate()
ax.bar(x,y)
plt.show()
```

# Category wise profit distribution

```python
plt.figure(figsize=(17,8))
sns.barplot(data['Category'],data['Profit'])
```

Technology sector earns most of the profit and furniture sector earns the lowest profit.

# Region wise sub-category classification

```python
plt.figure(figsize = (30, 12))
sns.countplot(x = 'Sub-Category', hue = 'Region', data = data, palette = 'magma')
plt.xticks(rotation = 45, fontsize = 20)
plt.yticks(fontsize = 20)
plt.xlabel('Sub-Categories', fontsize = 25)
plt.ylabel('', fontsize = 20)
plt.legend(loc = 5, fontsize = 20)
```

# Sub-category wise sales distribution

```python
fig,ax=plt.subplots(figsize=(30,12))
x=data['Sub-Category']
y=data['Sales']
ax.set_xlabel('Sub-Category')
ax.set_ylabel('Sales')
ax.set_title('Sub-Category wise Sales distribution')

fig.autofmt_xdate()
ax.bar(x,y)
plt.show()
```

# Sub-category wise profit distribution

```python
plt.figure(figsize=(17,8))
sns.barplot(data['Sub-Category'],data['Profit'])
```

Most of the profit earned by Copiers sector and least profit is earned by Fasteners.

# Correlation between different variables

```python
sns.heatmap(data.corr(),annot=True)
```

# East region is the weakest part in terms of sales and profit. Furniture sector has lowest profit and the lowest sales and if we look after sub-categories we will find fasteners has the lowest sales and earns the lowest profit. So, these are the weakest parts.
